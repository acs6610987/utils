```python
import requests
from bs4 import BeautifulSoup
import inspect
```

```python
cookie = {'Tequila_groups': 'vwyw8mbr5d1beifh1nobh0csodi6zw8w',
         'groups_lang': 'en',
         'petitpois': 'dismiss',
         '_ga': 'GA1.2.1801651510.1474895008',
         '_gid': 'GA1.2.1651566901.1516885917',
         '__utma': '33657086.1801651510.1474895008.1494228060.1504707099.23',
         '__utmz': '33657086.1504707099.23.9.utmcsr=google|utmccn=(organic)|utmcmd=organic|utmctr=(not%20provided)'}
response = requests.get("https://groups.epfl.ch/cgi-bin/groups/viewgroup?groupid=S13171", cookies=cookie)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
member_a_tags = soup.find(id='membersname').find_all('a')
member_a_tags
```

```python
member_links = list(map(lambda x: x['href'], member_a_tags))
member_links
```

```python
member_a_tags[0].get_text().strip()
```

```python
emails = []
domain = 'https://groups.epfl.ch'
for i in range(len(member_links)):
    url = domain + member_links[i]
    response = requests.get(url, cookies=cookie)
    member_soup = BeautifulSoup(response.text, 'html.parser')
    rows = member_soup.find(id='result-content').find_all('tr')
    for j in range(len(rows)):
        if rows[j].find('th').get_text().count('Email') >= 1:
            emails.append(rows[j].find('td').get_text().strip())
```

```python
emails
```

```python
for email in emails:
    print(email + ",")
```
